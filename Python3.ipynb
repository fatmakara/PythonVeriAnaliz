### 2021 EUR/TRY kur tablosunun incelenmesi

1. <i> Yahoo Finance</i> sitesinden 2021 yılına ait EUR/TRY kur değerlerini indiriniz

```python
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import matplotlib.pyplot as plt
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')
```

```python
kur_2021= yf.download('EURTRY=X', start ='2021-01-01', end='2021-12-31')
kur_2021
```

2. Veriyi özet metodları ile inceleyiniz

```python
kur_2021.head(10)
```

```python
kur_2021.info()
```

```python
kur_2021.describe()
```

3. Veri temizleme: <br>
	• Tabloda sadece "Adj Close" (düzeltilmiş kapanış) sütununu bırakınız.<br>
	• "Adj Close" sütununu " Kapanış (TRY)" olarak yeniden adlandırınız.

```python
kur_2021 = kur_2021[['Adj Close']]
kur_2021
```

```python
kur_2021.columns=['kapanış (TRY)']
kur_2021
```

4. Tabloya yeni sütun ekleme:<br>
	• Tabloya yeni bir sütun ekleyerek Kapanış (TRY) değerlerini 1 Ocak 2021 tarihindeki kapanış değerine göre normalize ediniz.<br>
	• BIST30 endeksinin 2021 yılı boyunca değişimini yorumlayınız.

```python
kur_2021 = kur_2021[kur_2021['kapanış (TRY)'] > 0] #warning almamak için
kur_2021['Normalize kapanış (TRY)'] = kur_2021['kapanış (TRY)'] / kur_2021.iloc[0, 0]
kur_2021
```

```python
b30_21 = yf.download('XU030.IS', start = '2021-01-01', end = '2022-01-01')
b30_21 = b30_21[['Adj Close']]
b30_21.columns = ['BIST30']
b30_21['Normalize BIST30'] = b30_21['BIST30'] / b30_21.iloc[0, 0]
b30_21
```

01-01-2021 tarihinde 1.000000 değerinde başlamış ve 30-12-2021 tarihinde 1.575236 değerinde ulaşarak %57 lik bir artış meydana gelmiştir.  Haziran ve temmuz ayında en düşük değerlere ulaşmıştır.


5. Veriyi grafik olarak inceleyiniz:<br>
	• Tabloyu px grafiği olarak görselleştiriniz.<br>
	• Tabloyu go. grafiği olarak görselleştiriniz.

```python
b30kur_2021= b30_21.merge(kur_2021, left_index = True, right_index = True)
#burada kapanış normalize 1'den başlamıyor.... 1.1.2021 değeri yok çünkü...
b30kur_2021['Normalize kapanış (TRY)'] = b30kur_2021['kapanış (TRY)'] / b30kur_2021.iloc[0, 2]
b30kur_2021
```

```python
#px grafiği
px.line(b30kur_2021, y = ['Normalize BIST30', 'Normalize kapanış (TRY)'],
        title='Normalize BIST 30 ve Normalize Kur (2021)',
        labels = dict({'Date':'Tarih', 'value': 'kapanış (TRY)'}),
        )
```

```python
#go grafigi

fig = go.Figure()
fig.add_trace(go.Bar(x = b30kur_2021.index, y = b30kur_2021['Normalize BIST30'], name = 'Normalize BIST30'))
fig.add_trace(go.Scatter(x = b30kur_2021.index, y = b30kur_2021['Normalize kapanış (TRY)'], name = 'Normalize EUR/TRY'))

fig.update_layout(title = 'Normalize BIST 30 ve Normalize Kur (2021)',
                  xaxis_title = 'Tarih', xaxis_dtick = 'M1',
                  yaxis_title = 'kapanış (TRY)')
```

### 2021 ve 2022 değerlerini birlikte inceleme

1. <i> Yahoo Finance</i> sitesinden 2022 BIST30 ve EUR/TRY kur değerlerini indiriniz

```python
indirme_listesi = ['XU030.IS', 'EURTRY=X']
b30eur22 = pd.DataFrame()

for i in indirme_listesi:
  data = yf.download(i, start = '2022-01-01', end = '2023-01-01')
  data = data['Adj Close'].rename(i)
  b30eur22 = pd.concat([b30eur22, data], axis = 1)

b30eur22
```

2. Indirdiğiniz verileri 2021 verilerini temizleyip düzenlediğiniz üzere hazırlayınız.

```python
b30eur22.columns = ['BIST30', 'Kapanış (TRY)'] #sütun isimlendirme
```

```python
b30eur22['Normalize BIST30'] = b30eur22['BIST30'] / b30kur_2021.iloc[0, 0] #dikkat paydada 2021 var
b30eur22['Normalize Kapanış (TRY)'] = b30eur22['Kapanış (TRY)'] / b30kur_2021.iloc[0, 2]  #dikkat paydada 2021 var
b30eur22 = b30eur22[b30eur22['BIST30'] > 0] #bist 30 olmayan satırlardan kurtulmak için
b30eur22
```

3. 2021 ve 2022 veri tablolarını tek bir tablo haline getiriniz.

```python
b30eur = pd.concat([b30kur_2021, b30eur22], axis = 0)
b30eur = b30eur[b30eur['BIST30'] > 0]
b30eur
```

4. Veriyi grafik olarak inceleyiniz:<br>
	• Tabloları px grafiği olarak görselleştiriniz.<br>
	• Tabloları go. grafiği olarak görselleştiriniz.<br>
	• Tabloları subplots grafiği olarak görselleştiriniz.

```python
#px grafigi
fig_3 = px.line(b30eur, y = ['Normalize BIST30', 'Normalize Kapanış (TRY)'])

fig_3.update_layout(title='Normalize BIST 30 ve Normalize Kur (2021-2022)',
                  xaxis_title = 'Tarih', xaxis_dtick = 'M1',
                  yaxis_title = 'Kapanış (TRY)')

fig_3.show()
```

```python
#go grafigi

fig = go.Figure()
fig.add_trace(go.Bar(x = b30eur.index, y = b30eur['Normalize BIST30'], name = 'Normalize BIST 30'))
fig.add_trace(go.Scatter(x = b30eur.index, y = b30eur['Normalize Kapanış (TRY)'], name = 'Normalize EUR/TRY'))

fig.update_layout(title = '2021-2022 Normalize BIST 30 ve Normalize Kur',
                  xaxis_title = 'Tarih', xaxis_dtick = 'M1',
                  yaxis_title = 'Kapanış (TRY)')
```
